```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/0xYUANTI/DarkTools/master/doc/darksouls-armor-stats-0.csv'
df = pd.read_csv(url)

# Drops 'url1' and 'location' columns
df.drop(['url1', 'location'], axis=1, inplace=True)

# Create a subset of resistances for analysis
subset = df[['physical', 'magic', 'fire', 'lightning', 'bleed', 'poison', 'curse']]

# Reverse the alphabetical order of the Y axis
subset.index = subset.index[::-1]

# Create a box and whisker plot for the subset with swapped axes
plt.figure(figsize=(10,8))
subset.boxplot(vert=False)
plt.title('Armor Stats')
plt.xlabel('Defense/Resistance')
plt.ylabel('Armor Class')
plt.show()
```
